## **교통 밀도에 따른 사고수**

```python
# prompt: accidents에 따라 traffic_density를 비교하는 그래프를 나타내줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/traffic_accidents_dict new.csv')

# 'accidents'에 따른 'traffic_density'의 평균을 계산하여 막대 그래프를 그립니다.
df_grouped = df.groupby('accidents')['traffic_density'].mean()
plt.figure(figsize=(10, 6))
df_grouped.plot(kind='bar')
plt.xlabel('Accidents')
plt.ylabel('Average Traffic Density')
plt.title('Average Traffic Density by Number of Accidents')
plt.show()

# 'accidents'와 'traffic_density' 사이의 상관관계를 계산하여 출력합니다.
correlation = df['accidents'].corr(df['traffic_density'])
print(f'Correlation between accidents and traffic density: {correlation}')

# 'accidents'와 'traffic_density'의 관계를 산점도로 표현합니다.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='accidents', y='traffic_density', data=df)
plt.xlabel('Accidents')
plt.ylabel('Traffic Density')
plt.title('Accidents vs. Traffic Density')
plt.show()
```

## **도시, 농촌에 따른 사고수**

```python
# prompt: accidents와 urban_area를 비교하는 그래프를 나타내줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/traffic_accidents_dict new.csv')

# 'accidents'와 'urban_area'의 관계를 산점도로 표현합니다.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='accidents', y='urban_area', data=df)
plt.xlabel('Accidents')
plt.ylabel('Urban Area')
plt.title('Accidents vs. Urban Area')
plt.show()

# 'accidents'에 따른 'urban_area'의 평균을 계산하여 막대 그래프를 그립니다.
df_grouped = df.groupby('accidents')['urban_area'].mean()
plt.figure(figsize=(10, 6))
df_grouped.plot(kind='bar')
plt.xlabel('Accidents')
plt.ylabel('Average Urban Area')
plt.title('Average Urban Area by Number of Accidents')
plt.show()

# 'accidents'와 'urban_area' 사이의 상관관계를 계산하여 출력합니다.
correlation = df['accidents'].corr(df['urban_area'])
print(f'Correlation between accidents and urban area: {correlation}')
```

## **교통사고의 요인들 전체 비교**

```python
# prompt: accidents
# traffic_fine_amount
# traffic_density
# traffic_lights
# pavement_quality
# urban_area
# average_speed
# rain_intensity
# vehicle_count
# time_of_day
# 표와 그래프로 나타내줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/traffic_accidents_dict new.csv')

# Display basic statistics for each column
print(df.describe())



# Visualize the relationship between accidents and other variables using scatter plots
plt.figure(figsize=(12, 8))

# Accidents vs. Traffic Fine Amount
plt.subplot(2, 3, 1)
plt.scatter(df['accidents'], df['traffic_fine_amount'])
plt.xlabel('Accidents')
plt.ylabel('Traffic Fine Amount')
plt.title('Accidents vs. Traffic Fine Amount')

# Accidents vs. Traffic Density
plt.subplot(2, 3, 2)
plt.scatter(df['accidents'], df['traffic_density'])
plt.xlabel('Accidents')
plt.ylabel('Traffic Density')
plt.title('Accidents vs. Traffic Density')

# Accidents vs. Traffic Lights
plt.subplot(2, 3, 3)
plt.scatter(df['accidents'], df['traffic_lights'])
plt.xlabel('Accidents')
plt.ylabel('Traffic Lights')
plt.title('Accidents vs. Traffic Lights')

# Accidents vs. Pavement Quality
plt.subplot(2, 3, 4)
plt.scatter(df['accidents'], df['pavement_quality'])
plt.xlabel('Accidents')
plt.ylabel('Pavement Quality')
plt.title('Accidents vs. Pavement Quality')

# Accidents vs. Average Speed
plt.subplot(2, 3, 5)
plt.scatter(df['accidents'], df['average_speed'])
plt.xlabel('Accidents')
plt.ylabel('Average Speed')
plt.title('Accidents vs. Average Speed')

# Accidents vs. Rain Intensity
plt.subplot(2, 3, 6)
plt.scatter(df['accidents'], df['rain_intensity'])
plt.xlabel('Accidents')
plt.ylabel('Rain Intensity')
plt.title('Accidents vs. Rain Intensity')

plt.tight_layout()
plt.show()


# You can also create more advanced visualizations using seaborn, such as pairplots
# sns.pairplot(df)
# plt.show()
```
